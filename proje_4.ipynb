```
from google.colab import drive
from IPython.display import clear_output
drive.mount('/content/drive')
metin = "/content/drive/MyDrive/şifrelenmemiş metin.txt"
sifrelenmisMetin = "/content/drive/MyDrive/şifrelenmiş metin.txt"
from os import path
def Sozluk_yukle():
  kelime=[]
  with open(metin) as openfile:
    for line in openfile:
      kelime.append(line.strip())
  return kelime


TurkishDictionary = Sozluk_yukle()

print(str(TurkishDictionary))


harfsayısı = 29
upper = ['A', 'B',	'C',	'Ç',	'D',	'E',	'F',	'G',	'Ğ',	'H',	'I',	'İ',	'J',	'K',	'L',	'M',	'N',	'O',	'Ö',	'P',	'R',	'S',	'Ş',	'T',	'U',	'Ü',	'V',	'Y',	'Z']
lower = ['a',	'b',	'c',	'ç',	'd',	'e',	'f',	'g',	'ğ',	'h',	'ı',	'i',	'j',	'k',	'l',	'm',	'n',	'o',	'ö',	'p',	'r',	's',	'ş',	't',	'u',	'ü',	'v',	'y',	'z']
def menugoster():
  print("---------------------.----")
  print("işlem seçiniz:")
  print("(1) dosyayı şifreleme")
  print("(2) şifreyi çözme")
  print("(4) çıkış\n")

def anahtar_al():
  anahtar = 0
  while True:
    anahtar_metin = input("Anahtar giriniz: ")
    try:           
      anahtar = int(anahtar_metin)
      return anahtar
    except:
      print("bu bir sayı değil..")
    
def Index_al(array, char):
  index = 0
  try:
    index = array.index(char)
    return index    
  except ValueError:
    return -1

def EncryptDecryptFile(inputFile, anahtar):
  dosya = open(inputFile, 'r') 
  sonuc = ""
  while True:
      char = dosya.read(1)
      if not char: 
          break 
      charIndex = Index_al(lower,char)
      if charIndex == -1:
          charIndex = Index_al(upper,char)
          if charIndex == -1: 
            sonuc = sonuc+ char
          else:
            encryptedChar = upper[(charIndex + anahtar) % harfsayısı]
            sonuc = sonuc + encryptedChar
      else:
        encryptedChar = lower[(charIndex + anahtar) % harfsayısı]
        sonuc = sonuc + encryptedChar      
           

  dosya.close()
  return sonuc


def dosyaya_yazma(text, fileName):
  çıktı = open(fileName, "wt")
  n = çıktı.write(text)
  çıktı.close()


def Sozluk_arama(word):

  if word.strip().lower() in TurkishDictionary:
    return True
  else:
    return False
  

while True:
  menugoster()
  choice = input("Hangi işlemi yapmak istersiniz:")
  if choice == "1":
    anahtar = anahtar_al()
    sonuc = EncryptDecryptFile(metin,anahtar)
    dosyaya_yazma(sonuc, sifrelenmisMetin)
    print("sonuç bu dosyada : şifrelenmiş metin.txt")

      
  elif choice == "2":
    anahtar = anahtar_al()
    anahtar = harfsayısı - anahtar
    sonuc = EncryptDecryptFile(sifrelenmisMetin,anahtar)
    dosyaya_yazma(sonuc, metin)
    print("sonuç bu dosyada: decrypted.txt")
    
  elif choice == "4":
    print("Çıkış yapılıyor..")
    break
  else:
    print("Böyle bir seçenek yok. Menüden seçiniz..")
```
